```python
class OddNumbersFilter:
    def __init__(self, numbers):
        self.numbers = numbers

    def filter(self):
        return [num for num in self.numbers if num % 2 != 0]

    @property
    def odd_numbers(self):
        return self.filter()

    def __str__(self):
        return "Ім'я: Олексій Група: КН-32"

числа = [
    [1, 2, 3, 4, 5, 6],
    [10, 21, 32, 43, 54],
    [7, 11, 14, 19, 22, 25] ]

print(f"{'='*30}")
for i, data in enumerate(числа, 1):
    obj = OddNumbersFilter(data)
    
    print(f"Інформація про автора: {obj}")
    print(f"Вхідні дані: {data}")
    print(f"Непарні числа (через property): {obj.odd_numbers}")
    print("-" * 30)

```
