```python
class OddNumbersFilter:
    def __init__(self, numbers: list[int]):
        # Зберігаємо список чисел
        self._numbers = numbers
        # Відразу готуємо список непарних чисел
        self._odd_numbers = [n for n in numbers if n % 2 != 0]

    def filter(self) -> list[int]:
        """Фільтрує та повертає лише непарні числа"""
        return self._odd_numbers

    @property
    def odd_numbers(self) -> list[int]:
        """Property, що повертає відфільтрований список непарних чисел"""
        return self._odd_numbers
    
    
```
